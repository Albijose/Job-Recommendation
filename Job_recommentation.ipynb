### User input preprocessing

```python
def user_input_preprocess(user_input):
  user_input['experience']=float(user_input['experience'])
  user_input['salary']=float(user_input['salary'])
  combined_user_features= f"{user_input['skills']} {user_input['work_type']} {user_input['location']} {user_input['education']}"
  user_input['combined_user_features'] = combined_user_features.lower()
  return user_input
```

```python
def vectorize_user_input(user_input, vectorizer):
    user_tfidf = vectorizer.transform([user_input['combined_user_features']])
    return user_tfidf
```

## Calculating cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_similar_jobs(user_tfidf, tfidf_matrix):
    # Calculate cosine similarity
    cosine_sim = cosine_similarity(user_tfidf, tfidf_matrix)
    return cosine_sim.flatten()
```

### filter jobs

```python
def filter_jobs(data, user_input,cosine_sim):
    filtered_data = data[
        (data['min_experience'] <= user_input['experience']) &
        (data['max_experience'] >= user_input['experience']) &
        (data['min_salary'] <= user_input['salary']) &
        (data['max_salary'] >= user_input['salary'])
    ]
    if 'work_type' in user_input and 'Work Type' in filtered_data.columns:
        filtered_data = filtered_data[filtered_data['Work Type'].str.lower() == user_input['work_type'].lower()]

    if 'location' in user_input and 'location' in filtered_data.columns:
        filtered_data = filtered_data[filtered_data['location'].str.lower() == user_input['location'].lower()]

    if filtered_data.empty:
        return pd.DataFrame(columns=data.columns.tolist() + ['similarity'])

    similarity_scores = cosine_sim.flatten()
    filtered_data = filtered_data.copy()
    filtered_data['similarity'] = similarity_scores[filtered_data.index]

    top_jobs = filtered_data.sort_values(by='similarity', ascending=False).head(3)

    return top_jobs
```

```python
def recommend_jobs(user_input, data, vectorizer, tfidf_matrix):
    processed_input = user_input_preprocess(user_input)
    user_tfidf = vectorize_user_input(processed_input, vectorizer)
    cosine_sim = cosine_similarity(user_tfidf, tfidf_matrix)

    top_jobs = filter_jobs(data, processed_input,cosine_sim)

    return top_jobs

```

```python
user_input = {
    "experience": 7,
    "location": "New York",
    "skills": "Python, Machine Learning, Data Analysis",
    "salary": 85000,
    "work_type": "Full-time",
    "education": "Bachelor's Degree"
}
recommended_jobs = recommend_jobs(user_input, data, vectorizer, tfidf_matrix)
print(recommended_jobs[['Job Title', 'Company','similarity']])
```
